```python
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv('../LC/Lending_Club_Accepted_2014_2018.csv')
df.head()
```

```python
param_df = df[['loan_amnt', 'grade', 'annual_inc', 'int_rate', 'fico_range_low', 'fico_range_high', 'loan_status']]
```

```python
param_df = param_df[(param_df['loan_status'] == 'Fully Paid') | (param_df['loan_status'] == 'Current') | (param_df['loan_status'] == 'Charged Off')]
param_df
```

```python
((param_df.isnull().sum()/len(df))*100).sort_values(ascending=False)
```

```python
param_df.dtypes
```

```python
grade_convert = {'A' : 1,
                 'B' : 2,
                 'C' : 3,
                 'D' : 4,
                 'E' : 5,
                 'F' : 6,
                 'G' : 7}
```

```python
param_df['converted_grade'] = param_df['grade'].astype('str').apply(lambda x: grade_convert[x])
param_df
```

```python
param_df['loan_status'].value_counts()
```

```python
loan_status_convert = {
    'Fully Paid' : 1,
    'Current' : 1,
    'Charged Off' : 0
}
```

```python
param_df['converted_loan_status'] = param_df['loan_status'].apply(lambda x: loan_status_convert[x])
param_df
```

```python
param_df['fico_range_mean'] = (param_df['fico_range_low'] + param_df['fico_range_high']) / 2
param_df
```

```python
import sklearn
import sklearn.svm
```

```python
param_df['predict_loan_status'] = 0
param_df
```

```python
reg = sklearn.linear_model.LinearRegression()
reg.fit(param_df[['loan_amnt', 'converted_grade', 'annual_inc', 'fico_range_mean', 'int_rate']], param_df['converted_loan_status'])
```
