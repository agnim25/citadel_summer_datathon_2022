```python
loan_ratio_labels = ["Under " + "{:.3f}".format(2500 / 250000), 
                      "{:.3f}".format(2500 / 250000) + " to " + "{:.3f}".format(7999 / 99999),
                      "{:.3f}".format(7999 / 99999) + " to " + "{:.3f}".format(4999 / 59999),
                      "{:.3f}".format(4999 / 59999) + " to " + "{:.3f}".format(2500 / 25000),
                      "{:.3f}".format(2500 / 25000)  + " to " +  "{:.3f}".format(49999 / 299999),
                      "{:.3f}".format(50000 / 300000) + " to " +  "{:.3f}".format(99999 / 399999),
                      "{:.3f}".format(100000 / 400000)  + " to " +  "{:.3f}".format(299999 / 499999),
                      "{:.3f}".format(300000 / 500000)  + " to " +  "{:.3f}".format(799999 / 699999),
                      "{:.3f}".format(800000 / 700000)  + " to " +  "{:.3f}".format(2000000 / 999999),
                     "More than " +  "{:.3f}".format(2000000 / 1000000)]

loan_ratio_labels
```

```python
lc_loan_ratio = df_lc['loan_amnt'] / df_lc['annual_inc']

loan_ratio_lc = [len(df_lc[lc_loan_ratio < 2500 / 250000]), 
                   len(df_lc[(2500 / 250000 <= lc_loan_ratio) & (lc_loan_ratio < 7999 / 99999)]),
                   len(df_lc[(7999 / 99999 < lc_loan_ratio) & (lc_loan_ratio < 4999 / 59999)]),
                   len(df_lc[(4999 / 59999 < lc_loan_ratio) & (lc_loan_ratio < 2500 / 25000)]),
                   len(df_lc[(2500 / 25000 < lc_loan_ratio) & (lc_loan_ratio < 49999 / 299999)]),
                   len(df_lc[(50000 / 300000 < lc_loan_ratio) & (lc_loan_ratio < 99999 / 399999)]),
                   len(df_lc[(100000 / 400000 < lc_loan_ratio) & (lc_loan_ratio < 299999 / 499999)]),
                   len(df_lc[(300000 / 500000 < lc_loan_ratio) & (lc_loan_ratio < 799999 / 699999)]),
                   len(df_lc[(800000 / 700000 < lc_loan_ratio) & (lc_loan_ratio < 2000000 / 999999)]),
                   len(df_lc[lc_loan_ratio >= 2000000 / 1000000])]

total_lc_loan = sum(loan_ratio_lc)

for i in range(len(loan_ratio_lc)):
    loan_ratio_lc[i] = loan_ratio_lc[i] / total_lc_loan * 100
    
loan_ratio_lc
```

```python
bank_loan_ratio = filtered_df_non_null['loan_amount'] / filtered_df_non_null['applicant_income']

loan_ratio_bank = [len(filtered_df_non_null[bank_loan_ratio < 2500 / 250000]), 
                   len(filtered_df_non_null[(2500 / 250000 <= bank_loan_ratio) & (bank_loan_ratio < 7999 / 99999)]),
                   len(filtered_df_non_null[(7999 / 99999 < bank_loan_ratio) & (bank_loan_ratio < 4999 / 59999)]),
                   len(filtered_df_non_null[(4999 / 59999 < bank_loan_ratio) & (bank_loan_ratio < 2500 / 25000)]),
                   len(filtered_df_non_null[(2500 / 25000 < bank_loan_ratio) & (bank_loan_ratio < 49999 / 299999)]),
                   len(filtered_df_non_null[(50000 / 300000 < bank_loan_ratio) & (bank_loan_ratio < 99999 / 399999)]),
                   len(filtered_df_non_null[(100000 / 400000 < bank_loan_ratio) & (bank_loan_ratio < 299999 / 499999)]),
                   len(filtered_df_non_null[(300000 / 500000 < bank_loan_ratio) & (bank_loan_ratio < 799999 / 699999)]),
                   len(filtered_df_non_null[(800000 / 700000 < bank_loan_ratio) & (bank_loan_ratio < 2000000 / 999999)]),
                   len(filtered_df_non_null[bank_loan_ratio >= 2000000 / 1000000])]

total_bank_loan = sum(loan_ratio_bank)

for i in range(len(loan_ratio_bank)):
    loan_ratio_bank[i] = loan_ratio_bank[i] / sum(loan_ratio_bank) * 100
    
loan_ratio_bank
```
