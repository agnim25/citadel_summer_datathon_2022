```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
frames = [pd.read_csv('../Bank/Bank_2014.csv'), pd.read_csv('../Bank/Bank_2015.csv'),
          pd.read_csv('../Bank/Bank_2016.csv'), pd.read_csv('../Bank/Bank_2017.csv')]
```

```python
df = pd.concat(frames)
df
```

```python
df.columns.to_list()
```

```python
filtered_df = df[['as_of_year', 'loan_amount_000s', 'applicant_income_000s']]
filtered_df
```

```python
((filtered_df.isnull().sum()/len(df))*100).sort_values(ascending=False)
```

```python
filtered_df_non_null = filtered_df.dropna()
```

```python
filtered_df_non_null.shape
```

```python
filtered_df_non_null['loan_amount_000s'] = filtered_df_non_null['loan_amount_000s'] * 1000
filtered_df_non_null['loan_amount_000s']
```

```python
filtered_df_non_null['applicant_income_000s'] = filtered_df_non_null['applicant_income_000s'] * 1000
filtered_df_non_null['applicant_income_000s']
```

```python
filtered_df_non_null.rename(columns={
    'applicant_income_000s' : 'applicant_income',
    'loan_amount_000s' : 'loan_amount'}, inplace=True)

filtered_df_non_null
```

```python
df_lc = pd.read_csv('../LC/Lending_Club_Accepted_2014_2018.csv')
df_lc.columns.to_list()
```

```python
df_lc['issue_d']
```

```python
df_lc['issue_d'] = pd.to_datetime(df_lc['issue_d'], format="%b-%Y").dt.to_period('M')
```

```python
df_lc['issue_d']
```

```python
import numpy as np
```

```python
bank_applicant_incomes = filtered_df_non_null['applicant_income']
bank_applicant_incomes = bank_applicant_incomes.to_numpy().astype('int')
```

```python
lc_applicant_incomes = df_lc['annual_inc']
lc_applicant_incomes = lc_applicant_incomes.to_numpy().astype('int')
```

```python
plt.style.use('seaborn-deep')
n_bins = 5
plt.hist([lc_applicant_incomes, bank_applicant_incomes], bins = n_bins,
         label=['LendingClub', 'Bank'], log=True)
plt.legend(loc='upper right')

plt.xlabel('Applicant Income')
plt.ylabel('Number of Occurences')

plt.title('Compare applicant income of people having loans at banks versus LendingClub')
plt.show()
```

```python
income_range_labels = ['Under 25000', '25000 to 59999', '60000 to 99999', '100000 to 199999', '200000 to 299999',
                 '300000 to 399999', '400000 to 499999', '500000 to 699999', '700000 to 999999', 'Over 1 million']
```

```python
income_range_lc = [len(df_lc[df_lc['annual_inc'] < 25000]), 
                   len(df_lc[(25000 < df_lc['annual_inc']) & (df_lc['annual_inc'] < 59999)]),
                   len(df_lc[(60000 < df_lc['annual_inc']) & (df_lc['annual_inc'] < 99999)]),
                   len(df_lc[(100000 < df_lc['annual_inc']) & (df_lc['annual_inc'] < 199999)]),
                   len(df_lc[(200000 < df_lc['annual_inc']) & (df_lc['annual_inc'] < 299999)]),
                   len(df_lc[(300000 < df_lc['annual_inc']) & (df_lc['annual_inc'] < 399999)]),
                   len(df_lc[(400000 < df_lc['annual_inc']) & (df_lc['annual_inc'] < 499999)]),
                   len(df_lc[(500000 < df_lc['annual_inc']) & (df_lc['annual_inc'] < 699999)]),
                   len(df_lc[(700000 < df_lc['annual_inc']) & (df_lc['annual_inc'] < 999999)]),
                   len(df_lc[df_lc['annual_inc'] > 1000000])]

income_range_lc
```

```python
income_range_bank = [len(filtered_df_non_null[filtered_df_non_null['applicant_income'] < 25000]), 
                   len(filtered_df_non_null[(25000 < filtered_df_non_null['applicant_income']) & (filtered_df_non_null['applicant_income'] < 59999)]),
                   len(filtered_df_non_null[(60000 < filtered_df_non_null['applicant_income']) & (filtered_df_non_null['applicant_income'] < 99999)]),
                   len(filtered_df_non_null[(100000 < filtered_df_non_null['applicant_income']) & (filtered_df_non_null['applicant_income'] < 199999)]),
                   len(filtered_df_non_null[(200000 < filtered_df_non_null['applicant_income']) & (filtered_df_non_null['applicant_income'] < 299999)]),
                   len(filtered_df_non_null[(300000 < filtered_df_non_null['applicant_income']) & (filtered_df_non_null['applicant_income'] < 399999)]),
                   len(filtered_df_non_null[(400000 < filtered_df_non_null['applicant_income']) & (filtered_df_non_null['applicant_income'] < 499999)]),
                   len(filtered_df_non_null[(500000 < filtered_df_non_null['applicant_income']) & (filtered_df_non_null['applicant_income'] < 699999)]),
                   len(filtered_df_non_null[(700000 < filtered_df_non_null['applicant_income']) & (filtered_df_non_null['applicant_income'] < 999999)]),
                   len(filtered_df_non_null[filtered_df_non_null['applicant_income'] > 1000000])]

income_range_bank
```

```python
total_bank = sum(income_range_bank)

for i in range(len(income_range_bank)):
    income_range_bank[i] = income_range_bank[i] / total_bank * 100
    
income_range_bank
```

```python
total_lc = sum(income_range_lc)

for i in range(len(income_range_lc)):
    income_range_lc[i] = income_range_lc[i] / total_lc * 100
    
income_range_lc
```

```python
plt.style.use('seaborn-deep')
n_bins = 5
plt.hist([income_range_lc, income_range_bank], bins = n_bins,
         label=['LendingClub', 'HMDA Dataset'], log=True)
plt.legend(loc='upper right')

plt.xlabel('Income Ranges (000s of Dollars)')
plt.ylabel('Percentage of loaners')

plt.title('Comparison of Debtor Income Distributions')
plt.show()
```
