```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../LC/Lending_Club_Accepted_2014_2018.csv')
df.head()
```

```python
df_60_month_term = df[df['term'] == ' 60 months']
df_60_month_term
```

```python
df_60_month_term['issue_d'] = pd.to_datetime(df_60_month_term['issue_d'], format="%b-%Y")
df_60_month_term['issue_d']
```

```python
df_60_month_term['grade'].unique()
```

```python
df_60_month_term_grade_A = df_60_month_term[df_60_month_term['grade'] == 'A']
df_60_month_term_grade_B = df_60_month_term[df_60_month_term['grade'] == 'B']
df_60_month_term_grade_C = df_60_month_term[df_60_month_term['grade'] == 'C']
df_60_month_term_grade_D = df_60_month_term[df_60_month_term['grade'] == 'D']
df_60_month_term_grade_E = df_60_month_term[df_60_month_term['grade'] == 'E']
df_60_month_term_grade_F = df_60_month_term[df_60_month_term['grade'] == 'F']
df_60_month_term_grade_G = df_60_month_term[df_60_month_term['grade'] == 'G']
```

```python
sorted_by_date_df_60_month_term_grade_A = df_60_month_term_grade_A.sort_values('issue_d')
sorted_by_date_df_60_month_term_grade_B = df_60_month_term_grade_B.sort_values('issue_d')
sorted_by_date_df_60_month_term_grade_C = df_60_month_term_grade_C.sort_values('issue_d')
sorted_by_date_df_60_month_term_grade_D = df_60_month_term_grade_D.sort_values('issue_d')
sorted_by_date_df_60_month_term_grade_E = df_60_month_term_grade_E.sort_values('issue_d')
sorted_by_date_df_60_month_term_grade_F = df_60_month_term_grade_F.sort_values('issue_d')
sorted_by_date_df_60_month_term_grade_G = df_60_month_term_grade_G.sort_values('issue_d')
```

```python
grade_A_ts = sorted_by_date_df_60_month_term_grade_A.groupby('issue_d')['int_rate'].mean()
grade_B_ts = sorted_by_date_df_60_month_term_grade_B.groupby('issue_d')['int_rate'].mean()
grade_C_ts = sorted_by_date_df_60_month_term_grade_C.groupby('issue_d')['int_rate'].mean()
grade_D_ts = sorted_by_date_df_60_month_term_grade_D.groupby('issue_d')['int_rate'].mean()
grade_E_ts = sorted_by_date_df_60_month_term_grade_E.groupby('issue_d')['int_rate'].mean()
grade_F_ts = sorted_by_date_df_60_month_term_grade_F.groupby('issue_d')['int_rate'].mean()
grade_G_ts = sorted_by_date_df_60_month_term_grade_G.groupby('issue_d')['int_rate'].mean()
```

```python
plt.figure(figsize=(16, 8), dpi=150)

grade_A_ts.plot(label='Group_A', color='red')
grade_B_ts.plot(label='Group_B', color='orange')
grade_C_ts.plot(label='Group_C', color='yellow')
grade_D_ts.plot(label='Group_D', color='green')
grade_E_ts.plot(label='Group_E', color='blue')
grade_F_ts.plot(label='Group_F', color='indigo')
grade_G_ts.plot(label='Group_G', color='violet')

plt.title('Average Interest Rate on Loans of Each Grade')
plt.xlabel('Time (every 3 months)')
plt.ylabel('Average Interest Rate (%)')
plt.legend()
```
